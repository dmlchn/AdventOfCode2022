```python
##Part 1

monkey_input = open('Day11Input.txt').read()
mod_input = monkey_input.split('\n')

##Create dictionary in which each monkey holds the items

#Figure out number of monkeys
monkeys = monkey_input.count('\n') // 7 + 1 
#Create dictionary
monkey_dict = {i: [int(num) for num in mod_input[1+ i*7][18:].split(',')] for i in range(monkeys)}

##Define function of what each monkey does during each step
def monkey_actions(monkey, item):
    divisor = int(mod_input[3+monkey*7][21:])
    true_receive_monkey = int(mod_input[4+monkey*7][29:])
    false_receive_monkey = int(mod_input[5+monkey*7][30:])
    #Inspection
    operation = mod_input[2+monkey*7].replace('old','item')[19:]
    item = eval(f'{operation}')
    #Get bored with item
    item = item // 3
    #Test to which monkey to throw to
    if item % divisor == 0:
        monkey_dict[true_receive_monkey].append(item)
    else:
        monkey_dict[false_receive_monkey].append(item)
    #Remove item from original monkey
    monkey_dict[monkey].pop(0)

monkey_inspections = {i: 0 for i in range(monkeys)}

def round():
    for monkey, items in monkey_dict.items():
        for item in items.copy():
            monkey_actions(monkey, item)
            monkey_inspections[monkey] += 1


for i in range(20):
    round()

dec_monkey_inspections = sorted(monkey_inspections.values(), reverse=True)
monkey_business_level = dec_monkey_inspections[0] * dec_monkey_inspections[1]
monkey_business_level
```
